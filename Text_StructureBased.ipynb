```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text = ("""
Space exploration has led to incredible scientific discoveries.From landing on the the moon to exploring mars,humanity continues to pushthe boundries
of what's possible beyond our planet.
these mission have not only expanded our knowledge of the universe but have also contributed to advancement in technology here on earth.Satellite communication ,GPS and
even certain medical imaging techniques trace their roots back to innovations drives by space programs"""
)
```

```python
#Initialize the splitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size =300,
    chunk_overlap=0,
)
```

```python
# perform the split
chunks = splitter.split_text(text)
```

```python
print(len(chunks))
print(chunks)
```
