```python
text_splitter = SemanticChunker(
    embeddings = HuggingFaceEmbeddings(),
    breakpoint_threshold_type="standard_deviation",
    breakpoint_threshold_amount=2
)
```

```python
docs = text_splitter.create_documents([sample])
```
